```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def d(n,x):
    if np.abs(x)>2/n:
        return 0
    elif (-2/n <= x) and (x<=-1/n):
        return (2*n+n**2*x)/3
    elif (-1/n<x) and (x<1/n):
        return n/3
    elif (1/n<=x) and (x<=2/n):
        return (2*n-n**2*x)/3
```

```python
def dvec(n,x):
    ans = np.zeros_like(x)
    for i in range(len(x)):
        ans[i] = d(n,x[i])
    return ans
```

```python
plt.plot(np.linspace(-10,10,1000),dvec(1,np.linspace(-10,10,1000)))
```

```python
xvals = np.linspace(-4,4,1000)
plt.figure(figsize=(5,3.5))
plt.xlim(-3,3)
plt.plot(xvals,dvec(1,xvals),'k-.',linewidth=2,label=r"$\delta_1(x)$")
plt.plot(xvals,dvec(5,xvals),'b--',linewidth=2,label=r"$\delta_{5}(x)$")
plt.plot(xvals,dvec(25,xvals),'r',linewidth=1,label=r"$\delta_{25}(x)$")
plt.legend()
plt.title("Sample Terms of Our Dirac Sequence")
plt.xlabel("x")
plt.ylabel("y")
plt.savefig("delta_sequence.pdf")
```
