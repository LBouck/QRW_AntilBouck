```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ten_steps = np.load("optimization5_10steps_0.5ic.npy")
twenty_steps = np.load("optimization5_20steps_0.5ic.npy")
thirty_steps = np.load("optimization5_30steps_0.5ic.npy")
```

```python
plt.style.use('seaborn-poster')
plt.rc('text', usetex=True)
#plt.rc('font', family='serif',serif='Computer Modern')
plt.rcParams['font.sans-serif']='Computer Modern Sans serif'
plt.rcParams['text.latex.preamble'] = [r'\usepackage[cm]{sfmath}']
fs = 16
plt.figure(figsize=(11,8))
plt.plot(np.abs(ten_steps[2,:]),"o")
plt.title(r"Gradient vs Iterations",fontsize=fs)
plt.xlabel(r"Iterations",fontsize=fs)
plt.ylabel(r"Gradient",fontsize=fs)
plt.savefig('extreems_opt5_steps10grad.pdf',format='pdf')
plt.show()

plt.figure(figsize=(11,8))
plt.semilogy(ten_steps[1,:],"o")
plt.title(r"Fit Error vs Iterations",fontsize=fs)
plt.tick_params(axis='y', which='minor')
plt.xlabel(r"Iterations",fontsize=fs)
plt.ylabel(r"Fit Error",fontsize=fs)
plt.ylim([np.min(ten_steps[1,:])-.0000001,np.max(ten_steps[1,:])+.0000001])
plt.savefig('extreems_opt5_steps10err.pdf',format='pdf')
plt.show()

plt.figure(figsize=(11,8))
plt.semilogy(ten_steps[0,:],ten_steps[1,:],"o")
plt.title(r"Fit Error vs Alpha",fontsize=fs)
plt.tick_params(axis='y', which='minor')
plt.xlabel(r"Alpha",fontsize=fs)
plt.ylabel(r"Fit Error",fontsize=fs)
plt.ylim([np.min(ten_steps[1,:])-.0000001,np.max(ten_steps[1,:])+.0000001])
plt.savefig('extreems_opt5_steps10errva.pdf',format='pdf')


plt.show()

plt.figure(figsize=(11,8))
plt.plot(ten_steps[0,:],"o")
plt.title(r"Alpha vs Iterations",fontsize=fs)
plt.xlabel(r"Iterations",fontsize=fs)
plt.ylabel(r"Alpha",fontsize=fs)
#plt.ylim([np.min(ten_steps[0,:]),np.max(ten_steps[0,:])])
plt.savefig('extreems_opt5_steps10alpha.pdf',format='pdf')
plt.show()
```
