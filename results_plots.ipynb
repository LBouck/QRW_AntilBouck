```python
import numpy as np
import matplotlib.pyplot as plt
from solvefokkerplanck_functions import *
%matplotlib inline
def blanchard_true(t,x):
    N=1/(np.sqrt(np.pi*t)*(2**(.25)))
    return N*np.exp(-np.sqrt(2)*t)*np.cosh(2*x)*np.exp(-(x**2)/(np.sqrt(2)*t))
def contcumprob(array,stepsize):
    answer=np.zeros(len(array))
    for i in range(1,len(answer)):
        answer[i]=stepsize*(array[i]+array[i-1])/2+answer[i-1]
    return answer
xlim=200
x_steps = 400
xvals=np.linspace(-xlim,xlim,x_steps)
plt.plot(xvals,contcumprob(blanchard_true(100,xvals),2*xlim/x_steps))
```
